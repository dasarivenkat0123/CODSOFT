```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Titanic-Dataset.csv')
survived = df['Survived'].sum()
print(f"Number of people survived: {survived}")
age_group = df.groupby('Age')['Survived'].count()
print(f"Age group:\n{age_group}")
gender_group = df.groupby('Sex')['Survived'].count()
print(f"Gender group:\n{gender_group}")
class_group = df.groupby('Pclass')['Survived'].count()
print(f"Ticket class group:\n{class_group}")
fare_group = df.groupby('Fare')['Survived'].count()
print(f"Fare group:\n{fare_group}")
cabin_group = df.groupby('Cabin')['Survived'].count()
print(f"Cabin group:\n{cabin_group}")
plt.bar(age_group.index, age_group.values)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Group with Frequency")
plt.show()
survival_rate_gender = df.groupby('Sex')['Survived'].mean() * 100
plt.bar(survival_rate_gender.index, survival_rate_gender.values)
plt.xlabel("Gender")
plt.ylabel("Survival Rate (%)")
plt.title("Survival Rate by Gender")
plt.show()
survival_rate_class = df.groupby('Pclass')['Survived'].mean() * 100
plt.bar(survival_rate_class.index, survival_rate_class.values)
plt.xlabel("Ticket Class")
plt.ylabel("Survival Rate (%)")
plt.title("Survival Rate by Ticket Class")
plt.show()
survival_rate_fare = df.groupby('Fare')['Survived'].mean() * 100
plt.bar(survival_rate_fare.index, survival_rate_fare.values)
plt.xlabel("Fare")
plt.ylabel("Survival Rate (%)")
plt.title("Survival Rate by Fare")
plt.show()
survival_rate_cabin = df.groupby('Cabin')['Survived'].mean() * 100
plt.bar(survival_rate_cabin.index, survival_rate_cabin.values)
plt.xlabel("Cabin")
plt.ylabel("Survival Rate (%)")
plt.title("Survival Rate by Cabin")
plt.show()
labels = ['Survived', 'Not Survived']
sizes = [survived, len(df) - survived]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Number of People Survived")
plt.show()
```
