```python
import matplotlib.pyplot as plt
import numpy as np
def weather_model(a, b, c, x):
    """Quadratic weather model function."""
    return a * x**2 + b * x + c
def read_from_file(filename):
    """Read weather coefficients and values from a file."""
    with open(filename, 'r') as file:
        lines = file.readlines()
    return [tuple(map(float, line.split())) for line in lines]
data_from_file = read_from_file('weather_data.txt')
plt.figure(figsize=(10, 8))
for idx, (a, b, c, x) in enumerate(data_from_file):
    x_values = np.linspace(-10, 10, 500)
    y_values = weather_model(a, b, c, x_values)
    plt.plot(x_values, y_values, label=f'Model {idx + 1}: {a}x² + {b}x + {c}')
    specific_y = weather_model(a, b, c, x)
    plt.scatter([x], [specific_y], color='red', label=f'Point {idx + 1} at x={x}, y={specific_y:.2f}')
plt.title("Weather Model Visualization")
plt.xlabel("x")
plt.ylabel("Weather Model Output (y)")
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a, b, c = 2, -4, 5
x = 5
def quadratic(a, b, c, x):
    return a * x**2 + b * x + c
x_values = np.linspace(-10, 10, 500)
y_values = quadratic(a, b, c, x_values)
specific_y = quadratic(a, b, c, x)
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=f'{a}x² + {b}x + {c}', color='blue')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.scatter([x], [specific_y], color='red', label=f'Point at x={x}, y={specific_y:.2f}')
plt.title("Graph of the Quadratic Equation")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```
