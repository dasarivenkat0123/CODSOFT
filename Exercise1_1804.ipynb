```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_model(a, b, c, x):
    return a * x**2 + b * x + c

a = 1.0
b = -2.0
c = 1.0


x_values = np.linspace(-10, 10, 400)
y_values = quadratic_model(a, b, c, x_values)

x_specific = 5  #
result = quadratic_model(a, b, c, x_specific)
print(f"The predicted weather variable for x={x_specific} is: {result}")


plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=f'y = {a}xÂ² + {b}x + {c}', color='blue')
plt.scatter(x_specific, result, color='red')
plt.title('Quadratic Model Visualization')
plt.xlabel('x values')
plt.ylabel('Predicted Weather Variable (y)')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
def singleinput(a, b, c, x):
    """Quadratic weather model function."""
    return a * x**2 + b * x + c
def read_from_file(filename):
    """Read weather coefficients and values from a file."""
    with open(filename, 'r') as file:
        lines = file.readlines()
    return [tuple(map(float, line.split())) for line in lines]
data_from_file = read_from_file('singleinput.txt')
plt.figure(figsize=(10, 8))
for idx, (a, b, c, x) in enumerate(data_from_file):
    x_values = np.linspace(-10, 10, 500)
    y_values = singleinput(a, b, c, x_values)
    plt.plot(x_values, y_values, label=f'Model {idx + 1}: {a}xÂ² + {b}x + {c}')
    specific_y = singleinput(a, b, c, x)
    plt.scatter([x], [specific_y], color='red', label=f'Point {idx + 1} at x={x}, y={specific_y:.2f}')
plt.title("Weather Model Visualization")
plt.xlabel("x")
plt.ylabel("Weather Model Output (y)")
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Read multiple sets of coefficients from a file
with open("multi_input.txt", "r") as file:
    lines = file.readlines()

# Parse the coefficients
coefficients = []
for line in lines:
    coefficients.append([float(x) for x in line.strip().split()])

# Generate x values
x = np.linspace(-10, 10, 400)

# Iterate over each set of coefficients
for idx, (a, b, c) in enumerate(coefficients):
    # Calculate y values based on the quadratic equation
    y = a * x**2 + b * x + c

    # Plot the quadratic equation
    plt.plot(x, y, label=f"Set {idx+1}: y = {a}x^2 + {b}x + {c}")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Quadratic Equations for Multiple Sets of Coefficients")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to solve a quadratic equation
def solve_quadratic(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        return "No Real Roots"
    root1 = (-b + discriminant**0.5) / (2 * a)
    root2 = (-b - discriminant**0.5) / (2 * a)
    return root1, root2

# Function to safely handle user input
def safe_input(prompt, cast_func=float):
    while True:
        try:
            return cast_func(input(prompt))
        except ValueError:
            print("Invalid input. Please enter a valid number.")

# Main program
num_equations = safe_input("Enter the number of quadratic equations: ", int)
for i in range(num_equations):
    print(f"Equation {i+1}")
    a = safe_input("Enter the coefficient a: ")
    b = safe_input("Enter the coefficient b: ")
    c = safe_input("Enter the coefficient c: ")

    roots = solve_quadratic(a, b, c)
    print(f"Equation {i+1} Roots:", roots)

    # Plotting the quadratic equation
    x = np.linspace(-10, 10, 400)
    y = a * x**2 + b * x + c
    plt.plot(x, y, label=f'Equation {i+1}: {a}x^2 + {b}x + {c}')

# Final plot adjustments
plt.axhline(0, color='black', linewidth=0.5)
plt.title("Quadratic Equation Plot - Debugged")
plt.legend()
plt.grid()
plt.show()
```
